## Oversampling??

```python
#Loading the dataset
df = pd.read_csv('diabetes_binary.csv')

df = df.astype('int') 
# rename Diabtes_012 to Diabetes
df.rename(columns={'Diabetes_binary': 'Diabetes'}, inplace=True)
```

```python
df.head()
```

```python
df.Diabetes.value_counts()
# imbalanced dataset
```

```python
# over sampling of the dataset to get a balanced dataset
class_0 = df[df['Diabetes'] == 0]
class_1 = df[df['Diabetes'] == 1]

# over sampling of the minority class 1
class_1_over = class_1.sample(len(class_0), replace=True)

# Creating a new dataframe with over sampled class 1 df and class 0 df
df_new = pd.concat([class_1_over, class_0], axis=0)

df_new.Diabetes.value_counts()

# plotting the new label distribution
#df_new['Diabetes'].value_counts().plot(kind='bar', title='Label Distribution after Oversampling')
```
